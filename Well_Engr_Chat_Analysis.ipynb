Define a function to convert WhatsApp text file into a pandas dataframe

```python
def parse_file(text_file):
    '''Convert WhatsApp chat log text file to a Pandas dataframe.'''
    
    # some regex to account for messages taking up multiple lines
    pat = re.compile(r'^(\d\d\/\d\d\/\d\d\d\d.*?)(?=^^\d\d\/\d\d\/\d\d\d\d|\Z)', re.S | re.M)
    with open(text_file, encoding = 'utf8') as f:
        data = [m.group(1).strip().replace('\n', ' ') for m in pat.finditer(f.read())]

    sender = []; message = []; datetime = []
    for row in data:

        # timestamp is before the first dash
        datetime.append(row.split(' - ')[0])

        # sender is between dash and colon
        try:
            s = re.search(' - (.*?):', row).group(1)
            sender.append(s)
        except:
            sender.append('')

        # message content is after the first colon
        try:
            message.append(row.split(': ', 1)[1])
        except:
            message.append('')

    df = pd.DataFrame(zip(datetime, sender, message), columns=['timestamp', 'sender', 'message'])
    df['timestamp'] = pd.to_datetime(df.timestamp, format='%d/%m/%Y, %H:%M')

    # remove events not associated with a sender
    df = df[df.sender != ''].reset_index(drop=True)
    
    return df
```

#### Emoji Usage by Engineers

```python
user_emoji_df = df.copy()
```

```python
user_emoji_df
```

```python
def emojis_list(x):
    emo_list = []
    found = r.findall(x)
    for emoji_found in found:
        emo_list.append(emoji_found)
    return emo_list
```

```python
user_emoji_df["list of emojis"] = user_emoji_df["message"].apply(lambda x : emojis_list(x))
```

```python
user_emoji_df
```

```python
def sender_emoji(sender):
    # filter the user_emoji_df to return only the rows that match a particular sender
    sender = user_emoji_df[user_emoji_df["sender"] == sender]
    
    # create an empty list which would be used to collate all emoji's sent by this sender
    sender_list = []
    
    # iterate through the filtered dataframe to collect all emoji's into the sender_list
    for item in sender["list of emojis"]:
        if len(item) != 0:
            sender_list += item
            
    # create an empty dictionary to be used in aggregating the emoji count       
    sender_dict = {}
    
    for emoji in sender_list:
        if emoji in sender_dict:
            sender_dict[emoji] += 1
        else:
            sender_dict[emoji] = 1
    
    # convert dictionary into pandas dataframe
    sender_emoji_df = pd.DataFrame(sender_dict, index=['Count']).T.sort_values(by='Count', ascending=False)
    
    return sender_emoji_df
```

#### Distinct conversations within the group

Let's define a conversation as a stream of messages where the gap between any two messages is not longer than 20 minutes.

```python
converse_df = df.copy()
```

```python
converse_df.head(3)
```

```python
# find time difference between current and previous message
converse_df['reply_time'] = (converse_df.timestamp.shift(-1) - converse_df.timestamp).apply(lambda x: x.total_seconds()/60).fillna(np.inf)
```

```python
# if reply time is more than 20 mins add 1 to counter
converse_df['conversation'] = (converse_df.reply_time > 20).cumsum().shift(1).fillna(0).astype(int) + 1
```

```python
df1 = converse_df.groupby('conversation').agg({'timestamp': ['min', 'max', 'count'], 
                                      'sender': ['first', 'unique', 'nunique']})
```

```python
df1['duration'] = (df1['timestamp']['max'] - df1['timestamp']['min']).apply(lambda x: x.total_seconds()/60)
```
